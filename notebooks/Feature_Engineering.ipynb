```python
import pandas as pd

# Load preprocessed data
df = pd.read_csv(r'C:\Users\HP\Desktop\Incident-Response-Dashboard\data\preprocessed_incident_data.csv')
```

```python
df.head()
```

```python
# Rolling average for severity
df['severity_rolling_avg'] = df['severity'].rolling(window=3).mean()
```

```python
# Lag features
df['lag_1_attack_type'] = df['attack_type_encoded'].shift(1)
df['lag_2_attack_type'] = df['attack_type_encoded'].shift(2)
```

```python
# Fill missing values after shifting
df.fillna(method='bfill', inplace=True)
```

```python
# Drop unnecessary columns
df = df.drop(columns=[ 'Unnamed: 0.1', 'Unnamed: 0'])
```

```python
# Save the feature-engineered data
df.to_csv(r'C:\Users\HP\Desktop\Incident-Response-Dashboard\data\feature_engineered_incident_data.csv', index=True, header=True, encoding='utf-8')

df.head()
```
