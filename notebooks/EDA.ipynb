Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of attack types
plt.figure(figsize=(10, 6))
sns.countplot(x='attack_type', data=df, palette='Set2')
plt.title("Distribution of Attack Types")
plt.xlabel("Attack Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of severity
plt.figure(figsize=(10, 6))
sns.histplot(df['severity'], bins=30, kde=True, color='red')
plt.title("Severity Distribution")
plt.xlabel("Severity")
plt.ylabel("Frequency")
plt.show()
```

```python
# Event type distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='event_type', data=df, palette='Set3')
plt.title("Distribution of Event Types")
plt.xlabel("Event Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Attacks over time (hourly)
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['hour'] = df['timestamp'].dt.hour
plt.figure(figsize=(10, 6))
sns.countplot(x='hour', data=df, palette='coolwarm')
plt.title("Incident Frequency by Hour of the Day")
plt.xlabel("Hour")
plt.ylabel("Incident Count")
plt.xticks(range(0, 24))
plt.show()
```

```python
# Severity over time (by hour)
df['severity'] = df['severity'].astype(float)
df['hour'] = df['timestamp'].dt.hour
plt.figure(figsize=(10, 6))
sns.boxplot(x='hour', y='severity', data=df, palette='coolwarm')
plt.title("Severity by Hour of the Day")
plt.xlabel("Hour")
plt.ylabel("Severity")
plt.xticks(range(0, 24))
plt.show()
```

```python
# Attack type severity correlation
plt.figure(figsize=(10, 6))
sns.boxplot(x='attack_type', y='severity', data=df, palette='Set1')
plt.title("Severity by Attack Type")
plt.xlabel("Attack Type")
plt.ylabel("Severity")
plt.xticks(rotation=45)
plt.show()
```
